# 카이제곱 검정

```python
from scipy.stats import chi2_contingency
```

## 1. Pricing_ab

```python
# 총 인원수
total_a = 429
total_b = 450

# 가입자 수 (플랜별 합)
subscribers_a = 44 + 38 + 38 + 38  # 합계
subscribers_b = 47 + 43 + 39 + 31  # 합계

# 2x2 표 생성
table1 = [
    [subscribers_a, total_a - subscribers_a],  # Variant A: 가입자, 비가입자
    [subscribers_b, total_b - subscribers_b]   # Variant B: 가입자, 비가입자
]

chi2, p, dof, expected = chi2_contingency(table1)

print("Chi-square Statistic:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("\nExpected Frequencies:\n", expected)
```

-> p-value > 0.05 이므로 귀무가설 **("Pricing A/B test 후 결과는 동일하다.")** 채택

## 2. Onboarding CTA

```python
# 총 인원수
total_original = 200
total_test = 207

# 가입자 수 (플랜별 합)
subscribers_original = 18 + 18 + 18 + 11  # 65명
subscribers_test = 23 + 18 + 17 + 12      # 70명

# 2x2 표 만들기 (가입자 vs 비가입자)
table2 = [
    [subscribers_original, total_original - subscribers_original],
    [subscribers_test, total_test - subscribers_test]
]

chi2, p, dof, expected = chi2_contingency(table2)

print("Chi-square Statistic:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("\nExpected Frequencies:\n", expected)
```

-> p-value > 0.05 이므로 귀무가설 **("Onboarding CTA A/B test 후 결과는 동일하다.")** 채택

## 3. Checkout design

### 이전 코드

```python
data2 = {
    'monthly_pro':  [0, 0],
    'monthly_basic': [0, 1],
    'annual_basic': [2, 5],
    'annual_pro':   [3, 0]
}

contingency_table3 = pd.DataFrame(data2, index=['Variant A', 'Variant B'])

# monthly_pro 제거
contingency_table3_reduced = contingency_table3.drop(columns=['monthly_pro'])

chi2_3, p_3, dof_3, expected_3 = chi2_contingency(contingency_table3_reduced)

print("Chi-square Statistic:", chi2_3)
print("p-value:", p_3)
print("Degrees of Freedom:", dof_3)
print("\nExpected Frequencies:\n", expected_3)
```

### 수정코드

```python
# 총 인원수
total_control = 60
total_redesign = 71

# 가입자 수 (플랜별 합)
subscribers_control = 3 + 2  # 5명
subscribers_redesign = 5 + 1 # 6명

# 2x2 표 만들기 (가입자 vs 비가입자)
table3 = [
    [subscribers_control, total_control - subscribers_control],
    [subscribers_redesign, total_redesign - subscribers_redesign]
]

chi2, p, dof, expected = chi2_contingency(table3)

print("Chi-square Statistic:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("\nExpected Frequencies:\n", expected)
```

-> p-value > 0.05 이므로 귀무가설 **("Checkout design A/B test 후 결과는 동일하다.")** 채택
